## Ciência de dados
</div>

Com o Colab, você pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A célula de códigos abaixo usa <strong>numpy</strong> para gerar dados aleatórios e <strong>matplotlib</strong> para visualizá-los. Para editar o código, basta clicar na célula e começar a editar.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import folium
from IPython.display import display
import numpy as np
from scipy.interpolate import splprep, splev

# Coordenadas dos pontos marcados
coordenadas = [
    (-22.96961589182126, -43.18186017722768),
    (-22.974594396385616, -43.186924188058704),
    (-22.977241621748178, -43.18151685445948),
    (-22.971512486610894, -43.175852028784085)
]

# Criando o mapa
mapa = folium.Map(location=[-22.9726, -43.1820], zoom_start=14)

# Adicionando marcadores
for coord in coordenadas:
    folium.Marker(location=coord).add_to(mapa)

# Coordenadas dos polígonos (mancha de óleo)
mancha_coords_outer = [
    (-22.9710, -43.1825),
    (-22.9715, -43.1818),
    (-22.9720, -43.1814),
    (-22.9725, -43.1809),
    (-22.9730, -43.1805),
    (-22.9735, -43.1807),
    (-22.9740, -43.1812),
    (-22.9745, -43.1817),
    (-22.9750, -43.1822),
    (-22.9747, -43.1827),
    (-22.9743, -43.1830),
    (-22.9737, -43.1832),
    (-22.9730, -43.1833),
    (-22.9723, -43.1831),
    (-22.9718, -43.1828)
]

mancha_coords_middle = [
    (-22.9715, -43.1822),
    (-22.9720, -43.1818),
    (-22.9725, -43.1814),
    (-22.9730, -43.1815),
    (-22.9735, -43.1817),
    (-22.9738, -43.1820),
    (-22.9740, -43.1825),
    (-22.9738, -43.1830),
    (-22.9733, -43.1833),
    (-22.9728, -43.1832),
    (-22.9723, -43.1828),
    (-22.9720, -43.1825),
    (-22.9718, -43.1822)
]

mancha_coords_inner = [
    (-22.9720, -43.1822),
    (-22.9725, -43.1818),
    (-22.9730, -43.1817),
    (-22.9735, -43.1820),
    (-22.9730, -43.1823),
    (-22.9725, -43.1823),
    (-22.9722, -43.1820)
]

# Função para suavizar os pontos do polígono
def smooth_polygon(coords, num_points=100):
    coords = np.array(coords)
    tck, u = splprep([coords[:, 0], coords[:, 1]], s=0.0, per=1)
    u_new = np.linspace(u.min(), u.max(), num_points)
    x_new, y_new = splev(u_new, tck, der=0)
    return list(zip(x_new, y_new))

# Suavizando as coordenadas dos polígonos
smoothed_coords_outer = smooth_polygon(mancha_coords_outer, num_points=100)
smoothed_coords_middle = smooth_polygon(mancha_coords_middle, num_points=100)
smoothed_coords_inner = smooth_polygon(mancha_coords_inner, num_points=100)

# Adicionando os polígonos suavizados ao mapa com o degradê existente
folium.Polygon(locations=smoothed_coords_outer, color=None, fill=True, fill_color='black', fill_opacity=0.4).add_to(mapa)
folium.Polygon(locations=smoothed_coords_middle, color=None, fill=True, fill_color='black', fill_opacity=0.6).add_to(mapa)
folium.Polygon(locations=smoothed_coords_inner, color=None, fill=True, fill_color='black', fill_opacity=0.8).add_to(mapa)

# Exibindo o mapa
display(mapa)
```
